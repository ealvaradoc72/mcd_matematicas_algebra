```python
#Ejercicio 01 de Algebra Lineal
#Autor : Enrique Alvarado

#Cargamos librerias
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def applyTransform(array_t, limits):
    #Primero definimos los puntos de mi circulo
    #Parametros de entrada
    #       array_t:  el arreglo a aplicar
    #       limits:   limites del grafico
    theta = np.linspace(0, 2*np.pi, 100)
    r = np.sqrt(1.0)
    x1 = r*np.cos(theta)
    x2 = r*np.sin(theta)
    
    fig, ax = plt.subplots(1)
    
    plt.xlim(limits[0])
    plt.ylim(limits[1])
    
    #aqui le aplicamos el arreglo y se transforma
    x11 = array_t[0,0]*x1 + array_t[0,1]*x2
    x21 = array_t[1,0]*x1 + array_t[1,1]*x2

    #Voy a necesitarlos en forma de vectores de coordenadas X y Y
    array_t = [array_t[0,0], array_t[1,0]],[array_t[0,1], array_t[1,1]]

    ax.plot(x11, x21)
    ax.set_aspect(1)
    for i in range(len(array_t)):
        x = np.concatenate([[0,0],array_t[i]])
        plt.quiver([x[0]],[x[1]],[x[2]],[x[3]],angles='xy', scale_units='xy', scale=1)
    plt.grid(linestyle='--')
    plt.gca().set_aspect('equal')
    plt.title('Ejercicio 1 Algebra Lineal', fontsize=8)

    plt.show()    
```

```python

#hacemos la descomposicion en valores singulares
A = [[1, 2],[2, 3]]
u, s, v = np.linalg.svd(A)

print ("A =", A)
print ("u =", u)
print ("s =", s)
print ("v =", v)
```

```python
#Se transforma con 'u'
limits = [[-1.25,1.25],[-1.25,1.25]]
applyTransform(u, limits)
```

```python
#Ahora se transformacon 's'
limits = [[-4.5,4.5],[-2,2]]
applyTransform(np.diag(s).dot(u), limits)
```

```python
#Con 'v'
limits = [[-4.5,4.5],[-4.5,4.5]]
applyTransform(u.dot(np.diag(s).dot(v)), limits)
```

```python
#Al aplicar 'A' directamente
limits = [[-4.5,4.5],[-4.5,4.5]]
applyTransform(np.array(A), limits)
```
